### Processamento dos Dados Textuais

#### Importando o Tokenizador do BERT

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased', from_pt=True)
```

#### Aplicando o Tokenizador 

```python
MAX_LENGTH = 32 # M√°ximo comprimento da senten√ßa
X_input_ids = np.zeros((len(df_train), MAX_LENGTH)) # Matriz para armazenamento dos ids
X_attn_masks = np.zeros((len(df_train), MAX_LENGTH)) # Matriz para armazenamento da "attention mask"
```

```python
# Tokenizando todas as senten√ßas com o "encode_plus" e retornando as matrizes de ids e "attention mask"
def preprocessing_dataset(df, ids, masks, tokenizer):
    for i, text in tqdm(enumerate(df_train['tweet_clean'])):
        tokenized_text = tokenizer.encode_plus(
            text,
            max_length=MAX_LENGTH, 
            truncation=True, 
            padding='max_length', 
            add_special_tokens=True,
            return_tensors='tf'
        )
        ids[i, :] = tokenized_text.input_ids
        masks[i, :] = tokenized_text.attention_mask
    return ids, masks
```

```python
# Aplicando a fun√ß√£o para tokeniza√ß√£o das senten√ßas contidas em "df_train"
X_input_ids, X_attn_masks = preprocessing_dataset(df_train, X_input_ids, X_attn_masks, tokenizer)
```

```python
print("Tamanho do vocabul√°rio ap√≥s a tokeniza√ß√£o: ", tokenizer.vocab_size)
```

#### Aplicando o One-Hot Encoding a coluna `label` do dataset

```python
# Transformando a coluna "label" em uma matriz de 0 e 1 que ser√° fornecida ao modelo
labels = pd.get_dummies(df_train['label']).to_numpy()
labels
```

```python
# Verifica√ß√£o da quantidade de labels diferentes e sua distribui√ß√£o percentual
values, counts = np.unique(labels, axis=0, return_counts=True)
print('Labels:', values)
print('Distribui√ß√£o (%):', np.around(100*(counts/len(labels)),2))
```

#### Composi√ß√£o e mapeamento do dataset

```python
# Composi√ß√£o do dataset que ser√° fornecido ao modelo, incluindo os ids, attention masks e labels
dataset = tf.data.Dataset.from_tensor_slices((X_input_ids, X_attn_masks, labels))
dataset.take(1) # tomando uma amostra dos dados
```

```python
# Fun√ß√£o de mapeamento para indicar como o modelo dever√° retornar a sa√≠da
def SentimentDatasetMapFunction(input_ids, attn_masks, labels):
    return {
        'input_ids': input_ids,
        'attention_mask': attn_masks
    }, labels
```

```python
# Aplicando fun√ß√£o de mapeamento
dataset = dataset.map(SentimentDatasetMapFunction)
dataset.take(1) # tomando uma amostra dos dados
```

```python
# Embaralhando o dataset e fixando o tamanho do "batch"
BATCH_SIZE = 16
BUFFER_SIZE = 10000

dataset = dataset.shuffle(buffer_size=BUFFER_SIZE, seed=SEED).batch(BATCH_SIZE, drop_remainder=True)
dataset.take(1) # tomando uma amostra dos dados
```

```python
# Dividindo o dataset de treino (df_train) em treino e valida√ß√£o

def get_dataset_partitions_tf(ds, ds_size, train_split=0.9, val_split=0.1):
    assert (train_split + val_split) == 1
    
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    
    train_ds = ds.take(train_size)    
    val_ds = ds.skip(train_size).take(val_size)
    
    return train_ds, val_ds, train_size, val_size

training_dataset, validation_dataset, train_size, val_size = get_dataset_partitions_tf(dataset, len(df_train))

print("Tamanho do conjunto de treinamento: ", train_size)
print("Tamanho do conjunto de valida√ß√£o: ", val_size)
```

## Composi√ß√£o do Modelo para An√°lise de Sentimentos com o BERT

### Importa√ß√£o do transformer e do modelo BERT pr√©-treinado no idioma Portugu√™s

```python
from transformers import TFBertModel
```

```python
model_bert = TFBertModel.from_pretrained('neuralmind/bert-base-portuguese-cased', from_pt=True)
```

### Composi√ß√£o do Modelo para An√°lise de Sentimentos

```python
def build_classifier_model():
    HIDDEN_LAYER_1_NODES=100
    DROPOUT_PROB=0.20
    N_CLASS = 3

    input_ids = tf.keras.layers.Input(shape=(MAX_LENGTH,), name='input_ids', dtype='int32')
    attn_masks = tf.keras.layers.Input(shape=(MAX_LENGTH,), name='attention_mask', dtype='int32')
    
    bert_embds = model_bert.bert(input_ids, attention_mask=attn_masks)[1] 
    intermediate_layer = tf.keras.layers.Dense(HIDDEN_LAYER_1_NODES, activation='relu', name='hidden_layer')(bert_embds)
    dropout = tf.keras.layers.Dropout(DROPOUT_PROB, name='dropout_1')(intermediate_layer) # (bert_embds)
    output_layer = tf.keras.layers.Dense(N_CLASS, activation='softmax', name='output_layer')(dropout)

    return tf.keras.Model(inputs=[input_ids, attn_masks], outputs=output_layer)

sentiment_model = build_classifier_model()
sentiment_model.summary()
```

#### Definindo o melhor Otimizador

**Fonte:** https://www.tensorflow.org/text/tutorials/classify_text_with_bert#evaluate_the_model

Para o ajuste fino, vamos usar o mesmo otimizador com o qual o BERT foi originalmente treinado: o "Adaptive Moments" (Adam). Este optimizador minimiza a perda de predi√ß√£o e faz regulariza√ß√£o por decaimento peso (sem uso de momentos), que tamb√©m √© conhecido como AdamW .

Para a taxa de aprendizagem ( init_lr ), voc√™ vai usar o mesmo esquema Bert-forma√ß√£o pr√©: Deteriora√ß√£o linear de uma taxa inicial de aprendizado te√≥rico, prefixado com uma fase de aquecimento linear durante os primeiros 10% de treinar os passos ( num_warmup_steps ). De acordo com o artigo de BERT, a taxa de aprendizado inicial √© menor para o ajuste fino (melhor de 5e-5, 3e-5, 2e-5).

#### Definindo a Fun√ß√£o de Perda e as M√©tricas

```python
from tensorflow_addons.metrics.f_scores import F1Score
```

```python
optim = tf.keras.optimizers.Adam(learning_rate=3e-5, decay=1e-6)
loss_func = tf.keras.losses.CategoricalCrossentropy()

precision = tf.keras.metrics.Precision(thresholds=0.5, top_k=None, class_id=None, name='precision', dtype=None)
recall = tf.keras.metrics.Recall(thresholds=0.5, top_k=None, class_id=None, name='recall', dtype=None)
f1 = F1Score(num_classes=3, average="weighted", threshold=0.5)

METRICS = [f1, 'categorical_accuracy', precision, recall]
```

#### Compilando e Treinando o Modelo para An√°lise de Sentimentos

```python
sentiment_model.compile(optimizer=optim, loss=loss_func, metrics=METRICS)
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='f1_score', # (ajustar para o conj. valida√ß√£o)
                                            mode='max', 
                                            patience=3, 
                                            restore_best_weights=True)

model_training = sentiment_model.fit(
    training_dataset,
    validation_data=validation_dataset,
    verbose='auto',
    class_weight=CLASS_WEIGHT,
    epochs=N_EPOCHS,
    callbacks=[callback]
)
```

### Classifica√ß√£o e Avalia√ß√£o com o Conjunto de Testes

```python
#tokenizer = BertTokenizer.from_pretrained('bert-base-cased')

def prepare_data(input_text, tokenizer):
    token = tokenizer.encode_plus(
        input_text,
        max_length=MAX_LENGTH, 
        truncation=True, 
        padding='max_length', 
        add_special_tokens=True,
        return_tensors='tf'
    )
    return {
        'input_ids': tf.cast(token.input_ids, tf.float64),
        'attention_mask': tf.cast(token.attention_mask, tf.float64)
    }

def make_prediction(model, processed_data, classes=[-1, 0, 1]):
    probs = model.predict(processed_data)[0]
    return classes[np.argmax(probs)]
```

```python
# For each class
precision = dict()
recall = dict()
thresholds = dict()
average_precision = dict()
n_classes = 3

for i in range(n_classes):
    precision[i], recall[i], thresholds[i] = precision_recall_curve(Y_test[:, i], y_score[:, i])
    average_precision[i] = average_precision_score(Y_test[:, i], y_score[:, i])
    
# A "micro-average": quantifying score on all classes jointly
precision['weighted'], recall['weighted'], thresholds['weighted'] = precision_recall_curve(
    Y_test.ravel(), y_score.ravel()
)
average_precision['weighted'] = average_precision_score(Y_test, y_score, average='weighted')
```
